## Problem Description - Daily Corona Confirm cases in India

The following code was to plot the daily confirm and recovered corona cases from India from 4th May 2020 to 17th April 2020

[Daily Confirmed Corona for India Data Source](https://www.covid19india.org/)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print('Hello Capstone Project Course!')
```

```python
db=pd.read_excel('Cummulative_Data.xlsx', index_col=None)
```

```python
plt.plot(db['Day'],db['I_Increase'], label='Daily Confirm Cases')
plt.plot(db['Day'],db['R_Increase'], label='Daily Removed (Recovered+Death) Cases')
plt.plot(db['Day'],db['I-R_Increase'], label='Daily Active Cases')
plt.legend(loc = 'upper left')
plt.xlim(0,45)
plt.xticks(rotation=90)
plt.xlabel("Day")
plt.ylabel("No of Cases")
```
